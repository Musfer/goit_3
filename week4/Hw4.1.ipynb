```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy import integrate
```

```python
time = np.linspace(0, 11, 12)
speed = np.array([25, 35, 45, 30, 60, 120, 100, 100, 70, 75, 80, 65])
print(time)
```

```python
plt.plot(time, speed)
plt.xlim([0,11])
plt.ylim([0,130])
plt.xlabel("time, h", fontsize=14)
plt.ylabel("speed, km/h", fontsize=14)
plt.grid()
```

```python
f_cubic = interp1d(time, speed, kind="cubic")
time_new = np.linspace(0, 11, 1000)
speed_cubic = f_cubic(time_new)
plt.plot(time, speed, linestyle="None", marker="o")
plt.plot(time_new, speed_cubic, color="red", label="speed  cubic interpolation")
plt.xlim([0,11])
plt.ylim([0,130])
plt.xlabel("time, h", fontsize=14)
plt.ylabel("speed, km/h", fontsize=14)
plt.legend()
plt.grid()
result = integrate.quad(f_cubic, 0, 11)
print(f"In qubic interbolation S = {result[0]} km")
```

```python
f_quadratic = interp1d(time, speed, kind="quadratic")
speed_quadratic = f_quadratic(time_new)
plt.plot(time, speed, linestyle="None", marker="o")
plt.plot(time_new, speed_new, color="red", label="speed  cubic interpolation")
plt.plot(time_new, speed_quadratic, color="green", label="speed quadratic interpolation")
plt.xlim([0,11])
plt.ylim([0,130])
plt.xlabel("time, h", fontsize=14)
plt.ylabel("speed, km/h", fontsize=14)
plt.legend()
plt.grid()
result = integrate.quad(f_quadratic, 0, 11)
print(f"In qubic interbolation S = {result[0]} km")
```
