# Problem 2 (Iris Fisher)

```python
import numpy as np
import matplotlib.pyplot as plt
import time

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC



```

### Let's compare three models (accuracy and execution time)

```python
model1 = SVC(kernel='linear')
model2 =  GaussianNB()
model3 =  KNeighborsClassifier(n_neighbors=5)

models = [
    {'name': 'SVC', 'model': model1},
    {'name': 'Naive Bayes', 'model': model2},
    {'name': 'KNN', 'model': model3},
]

# test set size array:
test_sizes = np.linspace(0.05, 0.4, 8)
```

### Load data set

```python
IF_data = load_iris()
```

```python
# execution time of each model
exec_time = []
# perforamnce results (accuracies)
train_acc = []
test_acc = []




for x in models:
    model = x.get('model')
    x['train_acc'] = []
    x['test_acc'] = []
    start_time = time.time()
    # repeating n times to average
    n = 100

    for test_size in test_sizes:
        train_accuracy = 0.0
        test_accuracy = 0.0
        for i in range(n):
            # spliting data into train/test sets
            X_train, X_test, y_train, y_test = train_test_split(
                IF_data.data,
                IF_data.target,
                test_size=test_size,
                shuffle=True,
                )
            model.fit(X_train, y_train)
            train_accuracy += accuracy_score(y_train, model.predict(X_train))/n
            test_accuracy += accuracy_score(y_test, model.predict(X_test))/n
        x['test_acc'].append(test_accuracy)
        x['train_acc'].append(train_accuracy)
        
    end_time = time.time()
    x['exec_time'] = "{:.2f} seconds".format(end_time - start_time)
#     x['exec_time'] = end_time - start_time
    
```

```python
# visualization of data
for i, model in enumerate(models):
    plt.plot(test_sizes, model.get('test_acc'),
            label=f"{model.get('name')}; {model.get('exec_time')}")

plt.xlabel("test set size")
plt.ylabel("test set accuracy")
plt.ylim(0.9,1)
plt.legend()

```
