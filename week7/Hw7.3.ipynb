# Problem 3 (Diabetes)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

from sklearn.linear_model import LinearRegression


```

```python
model = LinearRegression()

# test set size array:
test_sizes = np.linspace(0.05, 0.4, 8)
```

### Loadind dataset

```python
diabetes_data = load_diabetes()
```

```python
# perforamnce results (accuracies)
train_acc = []
test_acc = []

# repeating n times to average
n = 10

for test_size in test_sizes:
    test_accuracy = 0.0
    train_accuracy = 0.0
    for i in range(n):
        # spliting data into train/test sets
        X_train, X_test, y_train, y_test = train_test_split(
            diabetes_data.data,
            diabetes_data.target,
            test_size=test_size,
            shuffle=True,
            )
        model.fit(X_train, y_train)
        train_accuracy+= r2_score(y_train, model.predict(X_train))/n
        test_accuracy += r2_score(y_test, model.predict(X_test))/n
    train_acc.append(train_accuracy)
    test_acc.append(test_accuracy)



```

```python
plt.plot(test_sizes, train_acc, label="r2 score: train set")
plt.plot(test_sizes, test_acc, label="r2 score: test set")
plt.xlabel("test set size")
plt.ylabel("accuracy (r2 score)")
plt.legend()
```

# Conclusions:
### LinearRegression performs poorly on the load_diabetes dataset. 
### Changing the size of the training sample does not have a significant impact on the performance of the model.
