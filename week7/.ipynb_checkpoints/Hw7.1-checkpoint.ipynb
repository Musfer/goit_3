# Problem 1 (Wine)

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

```

## Tasks 1-4

```python
wine_data = load_wine()

X_train, X_test, y_train, y_test = train_test_split(
    wine_data.data,
    wine_data.target,
    test_size=0.25,
    shuffle=True,
)

model = SVC(kernel='linear')
model.fit(X_train, y_train)
train_accuracy = accuracy_score(model.predict(X_train),y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Train set accuracy = {train_accuracy}")
print(f"Test set accuracy  = {accuracy}")
```

# Testing with different test set sizes (Support Vector Machines (SVMs) model)

```python
model = SVC(kernel='linear')

# different test sizes
test_sizes = np.linspace(0.05, 0.3, 6)
# results
train_accuracies = []
test_accuracies = []

for test_size in test_sizes:
    # repeating n times to average our results
    n = 20
    train_accuracy = 0.0
    test_accuracy = 0.0
    for i in range(n):
        # spliting data into train/test sets
        X_train, X_test, y_train, y_test = train_test_split(
            wine_data.data,
            wine_data.target,
            test_size=test_size,
            shuffle=True,
            )
        model.fit(X_train, y_train)
        train_accuracy += accuracy_score(y_train, model.predict(X_train))/n
        test_accuracy += accuracy_score(y_test, model.predict(X_test))/n
    
    # filling results arrays 
    train_accuracies.append(train_accuracy)
    test_accuracies.append(test_accuracy)

plt.plot(test_sizes, train_accuracies, label=f"Accuracy on train set") 
plt.plot(test_sizes, test_accuracies, label=f"Accuracy on test set") 
plt.xlabel("test set size, %")
plt.ylabel("accuracy, %")
plt.ylim(0.9,1)
plt.legend()
    
```

## Let's try the same simulations but with Naive Bayes model
