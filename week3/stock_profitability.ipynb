```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "A": [25, -10, 10, 5, 35, 13],
    "B": [0, 15, -5, 5, 20, 25],
    "C": [10, 25, -15, -5, -5, 15]
})
df
```

```python
sns.regplot(x = "A",
            y = "B",
            data = df)
```

# B almost does not depend on A so they are independent

```python
sns.regplot(x = "B",
            y = "C",
            data = df)
```

```python
sns.regplot(x = "A",
            y = "C",
            data = df)
```
