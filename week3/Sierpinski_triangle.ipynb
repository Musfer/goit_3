```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# vertices of equilateral triangle
A = (0, 0)
B = (1, 0)
C = (0.5, np.sqrt(3)/2)
vertices = (A, B, C)

# chosing initial point inside the triangle
while True:
    P = (random.random(), random.random())
    if (P[1] < np.sqrt(3)*P[0]) and (P[1] < np.sqrt(3)*(1-P[0])):
        # P is inside of the triangle
        break
        
# ponts to plot
x = [A[0], B[0], C[0], P[0]] 
y = [A[1], B[1], C[1], P[1]]
for i in range(100000):
    # vertex to connect with last point
    D = random.choice(vertices)
    # append the new point 
    x.append((x[-1]+D[0])/2)
    y.append((y[-1]+D[1])/2)

plt.scatter(x,y, marker=".")
```
