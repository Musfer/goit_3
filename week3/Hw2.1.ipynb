# Зміна коефіцієнту народжуваності в Україні з часом 

```python

x = np.array([int(t) for t in df.columns[1:]])
y = np.array([float(t) for t in df.iloc[-1][1:]])
plt.plot(x, y, color='blue', linestyle='dashed', marker="o", mfc ="red")
plt.xlabel('рік')
plt.ylabel('Коефіцієнт народжуваності в Україні')
plt.show()
```

### 8. Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

# Порівняння динаміки народжуваності у декількох регіонах

```python
import random
# randint(0, 26)
rand_ids = random.sample(range(27), 5)
x = np.array([int(t) for t in df.columns[1:]])
for region_id in rand_ids:
    y = np.array([float(t) for t in df.iloc[region_id][1:]])
    plt.plot(x,y, label=f'{df.loc[region_id,"Регіон"]}', linestyle='-')
plt.xlabel('Рік')
plt.ylabel('Коефіцієнт народжуваності')
plt.legend()
plt.show()
```

### 9. Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
# new_df = df.loc[:,['Регіон', '2019']]
df[df['2019']>df['2019'].mean()].loc[:,['Регіон']]
```

### 10. У якому регіоні була найвища народжуваність у 2014 році?
